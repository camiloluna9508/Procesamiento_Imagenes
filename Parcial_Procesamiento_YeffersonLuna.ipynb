```python
from matplotlib.gridspec import GridSpec
from matplotlib import rc
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
```

```python
#Datos ax0
data1 = [10,20,30,40]
data2 = [12,22,32,42]
data3 = [8,18,28,45]
labels = ['2005', '2010','2015','2020']
promedio=[(data1[0]+data2[0]+data3[0])/3,(data1[1]+data2[1]+data3[1])/3,(data1[2]+data2[2]+data3[2])/3,(data1[3]+data2[3]+data3[3])/3]
#Datos ax1
x = np.linspace(0,2*np.pi,20)
y = np.sin(x)

#Datos ax2
x1 = np.linspace(0,10,100)
y1=np.log(x1)
y2=np.exp(x1)
#datos 
xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
```

```python
fig = plt.figure(figsize=(16,9))

gs = GridSpec(nrows=20,ncols=20)
ax0 = fig.add_subplot(gs[0:12,0:4])
wd = 0.6
x_pos = np.arange(1,2*len(labels),2)

ax0.barh(x_pos,data2,0.6,color='red',label="Audi")
ax0.barh(x_pos+wd,data1,0.6,color='green',label="BMW")
ax0.barh(x_pos+wd*2,data3,0.6,color='blue',label="Ferrari")
ax0.tick_params(axis='y',labelrotation=0)
ax0.set_yticks(x_pos+0.25)
ax0.set_ylabel('Ao')
ax0.set_yticklabels(labels)
ax0.set_xlabel('Carros')
ax0.set_title("Ventas Anuales")
ax0_1=ax0.twiny()
ax0_1.plot(promedio,x_pos, marker='.',linestyle='--', linewidth=0.7,color='orange',label="Promedio")
ax0.legend()
ax0_1.legend(loc = 'right')

ax1 = fig.add_subplot(gs[0:12,5:14])
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)

cp = ax1.contourf(X, Y, Z,cmap="jet")
ax1.set_title('z = sqrt(x^2 + y^2)')
fig.colorbar(cp,cmap="Reds")

ax1.set_ylabel('y')






ax2 = fig.add_subplot(gs[0:12,15:19])
ax2.plot(x1,y2,'b-',label="exp")
ax2.set_ylabel("e^x",color='black')
plt.yscale("log")

ax2_1 = ax2.twinx()
ax2_1.plot(x1,y1,"r",label="log")
ax2_1.set_ylabel("Log(x)",color="black")
ax2.legend()
ax2_1.legend(loc = 'lower right')

ax3 = fig.add_subplot(gs[14:,5:14])
ax3.plot(x,y,color='#3393FF', label='seno', marker='*',linestyle='-')
ax3.set_xlim([0,2*np.pi])
ax3.set_ylim([-1.1,1.1])
ax3.hlines(y = 1, xmin = 0, xmax = 2*np.pi, color ='black',linestyles="--")
ax3.hlines(y = -1, xmin = 0, xmax = 2*np.pi, color ='black',linestyles="--")
ax3.hlines(y = 0, xmin = 0, xmax = 2*np.pi, color ='black',linestyles="--")
ax3.set_xticks([0,2*np.pi])
ax3.set_xticklabels(["0","2"])
ax3.set_yticks([-1,0,1])
ax3.set_yticklabels(["-1","0","1"])
ax3.set_xlabel('Tiempo')
ax3.set_ylabel('amplitud')
ax3.set_title('Seno')
```
